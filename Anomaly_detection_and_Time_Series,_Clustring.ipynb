**Question 1 : What is Dimensionality Reduction? Why is it important in machine learning?**
Answer:

Dimensionality reduction is the process of reducing the number of input variables or features in a dataset while preserving as much important information as possible. Many datasets in machine learning contain hundreds or thousands of features, which increases complexity, storage requirements, computation time, and chances of overfitting. Dimensionality reduction transforms high-dimensional data into a lower-dimensional form, making it easier to visualize, analyze, and use in machine learning models.

It is important because it helps in removing noise and redundant features, improves model performance, speeds up training, reduces the risk of overfitting, and simplifies the overall machine learning pipeline. It enables better visualization of complex data, helps algorithms run faster, and often increases prediction accuracy by focusing only on the most essential features.

**Question 2: Name and briefly describe three common dimensionality reduction techniques.**

Answer:

1. Principal Component Analysis (PCA):
PCA is a statistical technique that transforms the original features into a new set of uncorrelated components called principal components. These components capture the maximum variance of the data, allowing most information to be preserved with fewer dimensions.

2. Linear Discriminant Analysis (LDA):
LDA reduces dimensions by finding a projection that maximizes class separability. It is mainly used in supervised learning when class labels are available, helping improve classification accuracy.

3. t-Distributed Stochastic Neighbor Embedding (t-SNE):
t-SNE is a nonlinear technique used to visualize high-dimensional data in 2D or 3D. It preserves local relationships between data points and is widely used in clustering and exploratory data analysis.

**Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.**

Answer:

Clustering is an unsupervised learning technique that groups similar data points together based on patterns or similarities within the data. Unlike supervised learning, clustering does not require labeled data. It helps identify hidden structures, relationships, and patterns among data points.

 - Three popular clustering algorithms are:

1. K-Means Clustering – Divides data into k clusters based on similarity and centroid distance.

2. Hierarchical Clustering – Builds a tree-like structure (dendrogram) to represent nested clusters.

3. DBSCAN (Density-Based Spatial Clustering of Applications with Noise) – Groups data points based on density and can identify noise and outliers effectively.

**Question 4: Explain the concept of anomaly detection and its significance.**

Answer:

 - Anomaly detection refers to the process of identifying unusual or abnormal patterns in data that do not follow expected behavior. These unusual patterns are called anomalies or outliers. Anomaly detection plays an important role in systems where deviations from normal behavior can indicate errors, fraud, failures, or threats.

 - Its significance lies in applications such as fraud detection in banking, intrusion detection in cybersecurity, fault detection in machinery, medical diagnosis, and quality control in manufacturing. Detecting anomalies early helps prevent losses, improves system reliability, ensures safety, and enhances decision-making.

** Question 5: List and briefly describe three types of anomaly detection techniques.**

Answer:

1. Statistical Techniques:
Use statistical models to define normal behavior (mean, variance, probability distribution). Any data point lying far outside the expected range is flagged as an anomaly.

2. Machine Learning-Based Techniques:
Algorithms such as Isolation Forest, One-Class SVM, and Autoencoders learn patterns of normal data and detect points that deviate significantly.

3. Distance or Density-Based Techniques:
Methods like k-Nearest Neighbors or DBSCAN identify anomalies based on distance or density. Points far from other points or in sparse regions are considered anomalies.

**Question 6: What is time series analysis? Mention two key components of time series data.**

Answer:

Time series analysis involves studying data points collected over time to understand patterns, trends, and behaviors. It is used for forecasting future values, detecting trends, and analyzing how data changes over time.

Two key components are:

1. Trend: The long-term upward or downward movement in the data.

2. Seasonality: Repeating patterns or cycles that occur at regular intervals (daily, monthly, yearly).

**Question 7: Describe the difference between seasonality and cyclic behavior in time series.**

Answer:

Seasonality refers to patterns that repeat at fixed and known intervals, such as daily temperature changes, monthly sales cycles, or yearly festival spikes. These are predictable and consistent over time.

Cyclic behavior involves irregular patterns that occur over longer, variable periods, such as economic cycles or business growth cycles. Unlike seasonality, cycles do not have a fixed duration and depend on external factors like economic conditions or market trends.


**Question 10: How can time series analysis be used for anomaly detection?**

Answer:

Time series analysis can detect anomalies by studying patterns such as trends, seasonality, and cyclic behavior over time. Anomalies occur when a data point significantly deviates from expected historical patterns. Techniques like ARIMA models, moving averages, exponential smoothing, and LSTM neural networks help forecast expected values. If actual values differ greatly from predicted values, they are flagged as anomalies.


Time series anomaly detection is widely used for detecting sudden spikes in network traffic, fraud in financial transactions, sudden drops in sensor readings, or unusual behavior in machines. It improves monitoring accuracy and enables timely detection of abnormal events.

```python
# Question 8: Write Python code to perform K-means clustering on a sample dataset.
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Create sample dataset
X, y = make_blobs(n_samples=300, centers=3, random_state=42)

# Apply K-Means
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Cluster labels
labels = kmeans.labels_

# Plotting results
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200)
plt.title("K-Means Clustering")
plt.show()
```

```python
# Question 9: What is inheritance in OOP? Provide a simple example in Python.
# Answer:
# Inheritance in Object-Oriented Programming (OOP) is the mechanism that allows one class (child class) to acquire properties and methods of another class (parent class). It promotes code reusability and reduces redundancy.
class Animal:
    def sound(self):
        return "Animal makes a sound"

class Dog(Animal):  # Dog inherits from Animal
    def sound(self):
        return "Dog barks"

# Using the classes
d = Dog()
print(d.sound())
```
